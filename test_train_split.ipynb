```python
import json
file_name=r"C:\Users\Rachel\PycharmProjects\npc-gzip\clause data\merged.json"
with open(file_name, 'r', encoding='utf-8') as f:
    lines = json.load(f)
```

```python
all_text=[]
all_classnames=[]
for line in lines:

        if len(line['annotations'][0]['result'])>0:
            text=line['data']['clause']
            if len(text)==0:
                continue
            classname=line['annotations'][0]['result'][0]['value']['choices'][0]


            text=text.replace('\n',' ')
            text=text.replace('\t',' ')
            text=text.replace('\\n',' ')
            l=text.split(" ")
            # print(classname)
            # print(" ".join(l))
            li=[]
            for i in l:
                if i=='':
                    pass
                else:
                    li.append(i)
            
            all_text.append(" ".join(li))
            all_classnames.append(classname)
```

```python
import pandas as pd

df=pd.DataFrame({'classname':all_classnames,"text":all_text})
```

```python
import pandas as pd
df = pd.read_csv('merged.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'class_column' is the name of the column representing classes
classes = df['classname'].unique()

train_data = pd.DataFrame()  # Initialize an empty DataFrame for training data
test_data = pd.DataFrame()   # Initialize an empty DataFrame for testing data

for class_val in classes:
    # Filter the data for each class
    class_data = df[df['classname'] == class_val]

    # Split the class data into 80% training and 20% testing
    train_class, test_class = train_test_split(class_data, test_size=0.2, random_state=42)

    # Concatenate the split class data to the respective train and test sets
    train_data = pd.concat([train_data, train_class])
    test_data = pd.concat([test_data, test_class])
```

```python
# Save the training and testing data to separate CSV files
train_data.to_csv('train_data.csv',header=False, index=False)
test_data.to_csv('test_data.csv', header=False, index=False)
```
